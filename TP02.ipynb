```python
import google.generativeai as genai

# Configurar a API Key do Gemini
API_KEY = "AIzaSyD46yC7LbBrLu_ytXyOUwqTNSht2gXn5Ac"
genai.configure(api_key=API_KEY)


# Texto de exemplo com 5.000 palavras
texto = " ".join(["palavra"] * 5000)

# Funo para calcular tokens com Gemini
def calcular_tokens_com_gemini(texto):
    """
    Usa o modelo Gemini para calcular o nmero de tokens em um texto.

    Args:
        texto (str): Texto completo para anlise.

    Returns:
        dict: Nmero de tokens usados no prompt e na resposta.
    """
    try:
        # Usar o modelo para calcular tokens
        response = genai.generate_text(
            model="text-bison-001",  # Modelo Gemini
            prompt=texto,
            max_output_tokens=10  # Limitar a sada para controlar o uso de tokens
        )
        # Retornar metadados de uso
        return response["usage_metadata"]
    except Exception as e:
        raise Exception(f"Erro ao processar o texto com o Gemini: {e}")

# Calcular e exibir tokens
try:
    tokens_info = calcular_tokens_com_gemini(texto)
    print(f"Tokens usados no prompt: {tokens_info['prompt_token_count']}")
    print(f"Tokens usados na resposta: {tokens_info['candidates_token_count']}")
    print(f"Total de tokens usados: {tokens_info['total_token_count']}")
except Exception as e:
    print(f"Erro ao calcular tokens: {str(e)}")
```

```python
import tiktoken

# Funo para estimar o nmero de tokens
def estimar_tokens(texto):
    """
    Estima o nmero de tokens de um texto usando o encoding 'cl100k_base'.

    Args:
        texto (str): Texto de entrada.

    Returns:
        list: Lista de tokens gerados.
    """
    # Inicializar o encoder com o esquema de codificao
    encoder = tiktoken.get_encoding("cl100k_base")
    # Tokenizar o texto
    tokens = encoder.encode(texto)
    return tokens

# Gerar um texto de exemplo com 5.000 palavras
texto_5000_palavras = " ".join(["palavra"] * 5000)

# Estimar tokens
tokens = estimar_tokens(texto_5000_palavras)

# Exibir resultados
print(f"Nmero total de tokens: {len(tokens)}")
print(f"Primeiros 20 tokens: {tokens[:20]}")
```

```python
import google.generativeai as genai
import os
import yaml
import json
import pandas as pd
import matplotlib.pyplot as plt

# Configurar a chave de API do Gemini
API_KEY = "AIzaSyD46yC7LbBrLu_ytXyOUwqTNSht2gXn5Ac"
genai.configure(api_key=API_KEY)

# Carregar a configurao do YAML
with open("roteiro.yaml", "r") as file:
    config = yaml.safe_load(file)

# Inicializar o modelo do Gemini
model = genai.GenerativeModel(config['roteiro']['llm_name'])

# Funo para ajustar e executar o prompt
# Funo para ajustar e validar o prompt
def obter_dados_turisticos(prompt):
    """
    Executa o prompt e valida se a resposta est no formato JSON esperado.

    Args:
        prompt (str): O prompt a ser executado.

    Returns:
        dict: Dados estruturados em JSON.
    """
    try:
        response = model.generate_content(prompt)
        print(f"Resposta bruta:\n{response.text}\n")  # Verificar o contedo retornado
        return json.loads(response.text)  # Tentar carregar como JSON
    except json.JSONDecodeError as e:
        print(f"Erro ao carregar resposta como JSON: {e}")
        return {}

dados_gerais = []
for cidade, prompt in config['roteiro']['prompts'].items():
    print(f"--- Gerando dados para {cidade.capitalize()} ---")
    dados = obter_dados_turisticos(prompt)
    for ponto, info in dados.items():
        info['nome'] = ponto  # Adicionar o nome como coluna separada
        dados_gerais.append(info)

# Verificar se os dados foram carregados corretamente
if not dados_gerais:
    print("Nenhum dado foi retornado. Verifique os prompts e as respostas do modelo.")
else:
    # Converter os dados em DataFrame
    df = pd.DataFrame(dados_gerais)

    # Exibir o DataFrame
    print("Dados Gerados:")
    print(df)

    # Configurar o grfico
    cores = {"Atenas": "blue", "Roma": "red"}
    df['cor'] = df['cidade'].map(cores)  # Mapear cores de acordo com a cidade

    # Plotar o grfico de barras horizontais
    plt.figure(figsize=(10, 6))
    plt.barh(df['nome'], df['qtd_visitas'], color=df['cor'])
    plt.xlabel("Quantidade de Visitantes (Milhes)")
    plt.ylabel("Pontos Tursticos")
    plt.title("Pontos Tursticos por Cidade")
    plt.legend(handles=[plt.Line2D([0], [0], color=cores['Atenas'], lw=4, label='Atenas'),
                        plt.Line2D([0], [0], color=cores['Roma'], lw=4, label='Roma')],
               title="Cidades")
    plt.show()
```

```python
import google.generativeai as genai
import os
import yaml
import json
import pandas as pd
import matplotlib.pyplot as plt
import re

# Configurar a chave de API do Gemini
API_KEY = "AIzaSyD46yC7LbBrLu_ytXyOUwqTNSht2gXn5Ac"
genai.configure(api_key=API_KEY)


# Carregar a configurao do YAML
with open("roteiro.yaml", "r") as file:
    config = yaml.safe_load(file)

# Inicializar o modelo do Gemini
model = genai.GenerativeModel(config['roteiro']['llm_name'])

# Funo para pr-processar e corrigir a resposta do modelo
def corrigir_json(resposta):
    """
    Limpa e tenta corrigir inconsistncias no JSON retornado pelo modelo.
    """
    try:
        # Captura apenas o bloco JSON entre as chaves mais externas
        resposta_corrigida = re.search(r"\{.*\}", resposta, re.DOTALL).group(0)
        # Remover aspas e texto no numrico de valores numricos
        resposta_corrigida = re.sub(r'"(\d+(?:\.\d+)?)\s*\(.*?\)"', r'\1', resposta_corrigida)
        return json.loads(resposta_corrigida)
    except Exception as e:
        print(f"Erro ao processar JSON corrigido: {e}")
        return {}

# Funo para executar e processar o prompt
def obter_dados_turisticos(prompt):
    try:
        response = model.generate_content(prompt)
        print(f"Resposta bruta:\n{response.text}\n")  # Verificar o contedo retornado
        return corrigir_json(response.text)
    except Exception as e:
        print(f"Erro ao executar o prompt: {e}")
        return {}

# Lista para acumular dados de ambas as cidades
dados_gerais = []
for cidade, prompt in config['roteiro']['prompts'].items():
    print(f"--- Gerando dados para {cidade.capitalize()} ---")
    dados = obter_dados_turisticos(prompt)
    if dados:
        for ponto, info in dados.items():
            info['nome'] = ponto
            dados_gerais.append(info)

# Verificar se os dados foram carregados corretamente
if not dados_gerais:
    print("Nenhum dado foi retornado. Verifique os prompts e as respostas do modelo.")
else:
    # Converter os dados em DataFrame
    df = pd.DataFrame(dados_gerais)

    # Corrigir a coluna de visitantes para nmeros inteiros
    df['qtd_visitas'] = pd.to_numeric(df['qtd_visitas'], errors='coerce')
    df = df.dropna(subset=['qtd_visitas'])  

    # Exibir o DataFrame
    print("Dados Gerados:")
    print(df)

    # Configurar o grfico
    cores = {"Atenas": "blue", "Roma": "red"}
    df['cor'] = df['cidade'].map(cores)  # Mapear cores de acordo com a cidade

    # Plotar o grfico de barras horizontais
    plt.figure(figsize=(10, 6))
    plt.barh(df['nome'], df['qtd_visitas'], color=df['cor'])
    plt.xlabel("Quantidade de Visitantes (Milhes)")
    plt.ylabel("Pontos Tursticos")
    plt.title("Pontos Tursticos por Cidade")
    plt.legend(handles=[plt.Line2D([0], [0], color=cores['Atenas'], lw=4, label='Atenas'),
                        plt.Line2D([0], [0], color=cores['Roma'], lw=4, label='Roma')],
               title="Cidades")
    plt.show()
```
